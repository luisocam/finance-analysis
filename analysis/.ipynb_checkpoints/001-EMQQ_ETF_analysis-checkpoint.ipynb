## Get ETF data
**We can also use a tool I made for this: https://github.com/luisocam/finance-analysis/blob/master/tools/get_IEX_data.py**

````python
ticker = 'EMQQ'
ticker_url = 'https://api.iextrading.com/1.0/stock/' + ticker + '/chart/5y'
ticker_response = requests.get(ticker_url)
ticker_data = ticker_response.json()
````

Pulling the data shows that we have just under 4 years of data available for analysis since the ETF's inception date was November 13, 2014.

````python
EMQQ_df = pd.DataFrame.from_dict(ticker_data)
EMQQ_df.set_index('date', inplace = True)
EMQQ_df.index = pd.to_datetime(EMQQ_df.index)

cols = ['open', 'high', 'low', 'close', 'volume']
EMQQ = pd.DataFrame(EMQQ_df, columns=cols)
EMQQ.head()
````

Save the extracted data in csv format to avoid having to request it again from IEX.

````python
EMQQ.to_csv('data/EMQQ_5Y.csv')
````

## Price evolution

We graph the ETF's daily closing price.

````python
plt.figure(figsize=(11,7))

ax = sns.lineplot(x=EMQQ.index, y=EMQQ.close)

ax.set_title('EMQQ - Emerging Markets Internet and Ecommerce ETF')
ax.set_ylabel('Price, USD')
ax.set_xlabel('');
````

## Daily returns
We first calculate the daily returns and store the result in a new column in our DataFrame.

````python
EMQQ['daily_return'] = EMQQ.close.pct_change()
EMQQ.head()
````

The graph below shows the daily returns over the study period. Consistent with the efficient market hypothesis, the price of EMQQ follows a random walk.

````python
plt.figure(figsize=(13,5))

ax = sns.lineplot(x=EMQQ.index, y=EMQQ.daily_return)

ax.set_title('EMQQ - Daily returns')
ax.set_ylabel('Return, %')
ax.set_xlabel('');
````

We calculate the mean and standard deviation of the daily returns for the study period.

````python
EMQQ_mean = np.mean(EMQQ.daily_return)
EMQQ_std = np.std(EMQQ.daily_return)
                  
print('Average return: ', f'{EMQQ_mean:.3%}')
print('Standard deviation: ', f'{EMQQ_std:.3%}')
````

We also plot a histogram of the daily returns over the past years. Daily returns for EMQQ look symmetric about the mean, however, the fat tails of the histogram show that there were some somewhat extreme results in the neighborhood of +/- 5%. This small [article](https://www.ifa.com/articles/with_stock_returns_normally_distributed/) explains it nicely.

````python
plt.figure(figsize=(10,7))

ax = sns.distplot(EMQQ.daily_return[1:])

ax.axvline(x=EMQQ_mean, color='k')
ax.axvline(x=EMQQ_mean + EMQQ_std, color='gray', linestyle='--')
ax.axvline(x=EMQQ_mean - EMQQ_std, color='gray', linestyle='--')

ax.set_title('EMQQ - Distribution of daily returns since Nov 13, 2014')
ax.set_ylabel('Frequency')
ax.set_xlabel('Daily Return')
ax.legend([f'Mean: {EMQQ_mean: .2%}', 
            f'SD 1: {EMQQ_mean + EMQQ_std: .2%}', f'SD 1: {EMQQ_mean - EMQQ_std: .2%}']);
````

## Volatility (daily, monthly and annualized)
We calulate the standard deviation pver the past 50 trading days in order to compute a measure for the daily, monthly and annualized volatility of EMQQ. We assume a year has [252 trading days](https://en.wikipedia.org/wiki/Volatility_(finance).

````python
volatility = np.std(EMQQ.daily_return[-50:])
volatility_monthly = volatility * np.sqrt(252 / 12)
volatility_annualized = volatility * np.sqrt(252)

print(f'EMQQ daily volatility: {volatility:.2%}')
print(f'EMQQ monthly volatility: {volatility_monthly:.2%}')
print(f'EMQQ annualized volatility: {volatility_annualized:.2%}')
````

## Autocorrelation
A common test of market efficiency is to check whether returns on day ```t``` affect the returns of the next day ```t + 1```.

````python
EMQQ['daily_returns_shifted'] = EMQQ.daily_return.shift(-1)
EMQQ[['daily_return', 'daily_returns_shifted']].corr()
````

Although autocorrelation is positive at 0.076 for the observed data, the figure below shows that there is no meaniingful tendency for returns on a given day to influence the returns on the following day, which is expected.

````python
plt.figure(figsize=(10,10))

ax = sns.scatterplot(x=EMQQ.daily_return, y=EMQQ.daily_returns_shifted)

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.set_title('EMQQ - Relationship between returns on successive days')
ax.set_ylabel('Return day t + 1, %')
ax.set_xlabel('Return day t, %');
````

## Performance (1M, 3M, 6M, YTD, 1Y, 3Y, MAX)

We create a function that lets us compute returs for different priods. It returns a dataframe that we cna also use for graphing purposes. Again, we assume a year has 252 trading days. These returns are also based solely on price, no distributions are considered.

````python
def total_return(period, current_year, dataframe):
    """Calculates the return for a given period and returns a dataframe for that period"""
    cols = ['daily_return']
    df = pd.DataFrame(dataframe, columns=cols)
    
    if period == '1M':
        df = df[int(-252 / 12):]
    elif period == '3M':
        df = df[int(-252 / 4):]
    elif period == '6M':
        df = df[int(-252 / 2):]
    elif period == 'YTD':
        df = df.loc[current_year]
    elif period == '1Y':
        df = df[int(-252):]
    elif period == '3Y':
        df = df[int(-252 * 3):]
    elif period == 'MAX':
        pass
    else:
        return print ('Please select a supported study period.')
    
    df['performance'] = df.daily_return.cumsum()
    
    if df.performance[-1] < 0:
        graph_color = 'red'
    else:
        graph_color = 'green'
    
    plt.figure(figsize=(11,5))
    
    plt.fill_between(df.index, df.performance, alpha=0.5, color=graph_color)
    
    plt.title(f'EMQQ {period} Performance: {df.performance[-1]:.2%}')
    plt.ylabel('Total return, %')
    
    return plt
````

### Performance graphs

````python
periods = ['1M', '3M', '6M', 'YTD', '1Y', '3Y', 'MAX']

for period in periods:
    total_return(period, '2018', EMQQ);
````
